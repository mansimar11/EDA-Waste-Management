```python
#  1. Waste Stress Score = How much of landfill is used per day
df['Waste Stress Score'] = df['Waste Generated (Tons/Day)'] / df['Landfill Capacity (Tons)']
```

```python
#  2. Recycling ROI = Efficiency of money spent on waste management
df['Recycling ROI'] = df['Recycling Rate (%)'] / df['Cost of Waste Management (₹/Ton)']
```

```python
# 💡 3. Infrastructure Strain Index = How much waste one efficiency point handles
df['Infra Strain Index'] = df['Waste Generated (Tons/Day)'] / df['Municipal Efficiency Score (1-10)']
```

```python
#  4. Awareness Effectiveness = Recycling Rate improvement per campaign
df['Awareness Effectiveness'] = df['Recycling Rate (%)'] / (df['Awareness Campaigns Count'] + 1)  # Avoid division by 0
```

```python
#  5. Cost Efficiency Index = Waste handled per ₹ spent
df['Cost Efficiency Index'] = df['Waste Generated (Tons/Day)'] / df['Cost of Waste Management (₹/Ton)']
```

```python
#  6. Urban Waste Pressure Index = Waste generated per population density unit
df['Urban Waste Pressure'] = df['Waste Generated (Tons/Day)'] / df['Population Density (People/km²)']
```

```python
# Expand display width and precision
pd.set_option('display.width', 150)
pd.set_option('display.max_columns', None)
pd.set_option('display.float_format', '{:.3f}'.format)
```

```python
# Group by City and compute average KPI values
grouped = df.groupby('City/District')[[
    'Recycling ROI',
    'Awareness Effectiveness',
    'Cost Efficiency Index',
    'Infra Strain Index',
    'Urban Waste Pressure'
]].mean().reset_index()

# Display top/bottom 5 for each KPI

print(" Top 5 Cities by Recycling ROI:")
print(grouped.sort_values('Recycling ROI', ascending=False).head())

print("\n Top 5 Cities by Awareness Effectiveness:")
print(grouped.sort_values('Awareness Effectiveness', ascending=False).head())

print("\n Top 5 Cities by Cost Efficiency Index:")
print(grouped.sort_values('Cost Efficiency Index', ascending=False).head())

print("\n Bottom 5 Cities by Infra Strain Index (high = more strain, worse):")
print(grouped.sort_values('Infra Strain Index').head())

print("\n Top 5 Cities by Urban Waste Pressure:")
print(grouped.sort_values('Urban Waste Pressure', ascending=False).head())
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Set wider display for better output readability
pd.set_option('display.width', 150)
pd.set_option('display.max_columns', None)
pd.set_option('display.float_format', '{:.3f}'.format)

# Group by city and calculate mean KPI values
grouped = df.groupby('City/District')[[
    'Recycling ROI',
    'Awareness Effectiveness',
    'Cost Efficiency Index',
    'Infra Strain Index',
    'Urban Waste Pressure'
]].mean().reset_index()

# Define KPIs to plot
kpi_list = {
    'Recycling ROI': 'Top 5 Cities by Recycling ROI',
    'Awareness Effectiveness': 'Top 5 Cities by Awareness Effectiveness',
    'Cost Efficiency Index': 'Top 5 Cities by Cost Efficiency Index',
    'Infra Strain Index': 'Bottom 5 Cities by Infra Strain (low = better)',
    'Urban Waste Pressure': 'Top 5 Cities by Urban Waste Pressure'
}

# Plotting all KPI charts
for kpi, title in kpi_list.items():
    if 'Bottom' in title:
        top5 = grouped.sort_values(kpi, ascending=True).head(5)
    else:
        top5 = grouped.sort_values(kpi, ascending=False).head(5)

    plt.figure(figsize=(10, 6))
    plt.barh(top5['City/District'], top5[kpi], color='steelblue')
    plt.xlabel(kpi)
    plt.title(title)
    plt.gca().invert_yaxis()
    plt.tight_layout()
    plt.show()
```

```python
# List of KPI columns to normalize
kpi_cols = ['Recycling ROI', 'Awareness Effectiveness', 'Cost Efficiency Index',
            'Infra Strain Index', 'Urban Waste Pressure']

# Normalize each column manually using min-max scaling
for col in kpi_cols:
    min_val = grouped[col].min()
    max_val = grouped[col].max()
    grouped[f'{col} (Norm)'] = (grouped[col] - min_val) / (max_val - min_val)
```

```python
# Invert the negative KPIs directly
grouped['Infra Strain (Inv)'] = 1 - grouped['Infra Strain Index (Norm)']
grouped['Urban Waste Pressure (Inv)'] = 1 - grouped['Urban Waste Pressure (Norm)']

# Calculate Composite Score by averaging all 5 KPI values (3 positive + 2 inverted)
grouped['Composite Score'] = (
    grouped['Recycling ROI (Norm)'] +
    grouped['Awareness Effectiveness (Norm)'] +
    grouped['Cost Efficiency Index (Norm)'] +
    grouped['Infra Strain (Inv)'] +
    grouped['Urban Waste Pressure (Inv)']
) / 5
```

```python
print("\n Top 5 Best Overall Cities (based on all KPIs):")
print(grouped.sort_values('Composite Score', ascending=False)[['City/District', 'Composite Score']].head())

print("\n Bottom 5 Worst Overall Cities (based on all KPIs):")
print(grouped.sort_values('Composite Score', ascending=True)[['City/District', 'Composite Score']].head())
```
